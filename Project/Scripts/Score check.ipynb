```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')


x = ['iteration','score']

df = pd.read_csv("resultHillClimbing.txt", names=x)

iteration = df.iteration.to_list()
score = df.score.to_list()

plt.plot(iteration,score)
plt.title('Wyniki w algorytmie HillClimbing')
plt.xlabel('iterations')
plt.ylabel('score')
plt.show()
```

```python
x2 = ['iteration','score']

df = pd.read_csv("resultGenetic.txt", names=x2)

iteration2 = df.iteration.to_list()
score2 = df.score.to_list()

plt.plot(iteration2,score2,'ro')
plt.title('Wyniki w algorytmie genetycznym')
plt.xlabel('iterations')
plt.ylabel('score')
plt.show()
```

```python
x3 = ['iteration','score']

df = pd.read_csv("resultRandomHillClimbing.txt", names=x3)

iteration3 = df.iteration.to_list()
score3 = df.score.to_list()

x4 = ['iteration','score']

df = pd.read_csv("resultTabu.txt", names=x4)

iteration4 = df.iteration.to_list()
score4 = df.score.to_list()

x5 = ['iteration','score']

df = pd.read_csv("resultSimulatedAnnealing.txt", names=x5)

iteration5 = df.iteration.to_list()
score5 = df.score.to_list()

plt.plot(iteration3,score3, label = "random")
plt.plot(iteration4,score4, label =" tabu")
plt.plot(iteration5,score5, label = "sa")
plt.title('Porwnanie dziaania algorytmw przy 1000 iteracji')
plt.xlabel('score')
plt.ylabel('iterations')
plt.legend()
plt.show()
```
