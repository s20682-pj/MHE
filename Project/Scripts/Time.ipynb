```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')


x = ['name','time']

df = pd.read_csv("result.txt", names=x)

name = df.name.to_list()
time = df.time.to_list()

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(name,time)
plt.title("Porównanie czasu wykonania algorytmów")
plt.show()


plt.show()
```

Ekspreryment został wykonany dla jak najbardziej zbliżonych warunków. Można zauważyć, że najszybciej wykonuje się algorytm wspinaczkowy, a najwolniej algorytm genetyczny. Jest to spowodowane tym, że algorytm genetyczny sprawdza największą ilość rozwiązań.
